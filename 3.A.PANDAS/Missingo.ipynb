```python
import pandas as pd 
import missingno as msno
```

```python
df = pd.read_csv('nba.csv')
```

```python
msno.matrix(df)
```

```python
# Visualize the number of missing 
# values as a bar chart 
msno.bar(df) 
```

```python
# Visualize the correlation between the number of 
# missing values in different columns as a heatmap 
msno.heatmap(df)
```
