```python
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.tail(3)
```

```python
df = df.drop(columns='Species')
```

```python
df.head(3)
```

```python
df['SepalWidthCm'].mean()
```

```python
#= 5.1 - 5.84 = (-0.73) 
```

```python
from sklearn.cluster import KMeans
```

```python
df.columns
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
    sse.append(km.inertia_)
    print(km.inertia_)
    
```

```python
sse
```

```python
import matplotlib.pyplot as plt
plt.plot(k_rng,sse)
```
